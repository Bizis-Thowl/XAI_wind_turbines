```python
#Reimport
ctgan = ctgan.load("./data/multi/model")

class_target_name = "Risk Level"

train = pd.read_csv("./data/multi/train_hydraulic.csv", sep=",")

drop_cols = ["Turbine_ID", "Timestamp", "Unnamed: 0", "index_y"]
train = train.drop(columns = drop_cols)

fill_method="ffill"
train = train.fillna(method=fill_method)

base = train
generated_same = sammples
generated_half = ctgan.sample(15000)
generated_double = ctgan.sample(60000)
generated_fact10 = ctgan.sample(120000)

test = pd.read_csv("./data/multi/test_hydraulic.csv", sep=",")
drop_cols = ["Turbine_ID", "Timestamp", "Unnamed: 0", "index_y"]
test = test.drop(columns = drop_cols)
fill_method="ffill"
test = test.fillna(method = fill_method)
test = test.fillna(method = "backfill")
```

```python
test.notna()
```

```python
def split_x_y(data:DataFrame ,target_name: str):
    drop = [target_name]
    x = data.drop(columns=target_name)
    y = data[target_name]
    return x,y

def checkNaN(data: DataFrame):
    re = str(data.isnull().sum().sum())
    #print(re)
    return re

def get_eval_scores(inputs_train:DataFrame, targets_train:DataFrame, inputs_test:DataFrame, targets_test:DataFrame):    
    dtr = DecisionTreeClassifier(random_state=0)
    dtr.fit(inputs_train, targets_train)
    eval_predict = dtr.predict(inputs_test)
    accuracy = balanced_accuracy_score(targets_test, eval_predict)
    return accuracy

def train_run(data: DataFrame, test:DataFrame):
    train_x, train_y = split_x_y(data, class_target_name)    
    test_x, test_y = split_x_y(test, class_target_name)
    return get_eval_scores(train_x, train_y, test_x, test_y)
    
```

```python
base_accuracy = train_run(base, test)
generated_fact10_accuracy = train_run(generated_fact10, test)
combined = pd.concat([generated_fact10, base])
combined_accuracy = train_run(combined,test)
print("eBas:" + str(base_accuracy))
print("Generated_Only:" + str(generated_fact10_accuracy))
print("Combined:" + str(combined_accuracy))
```
